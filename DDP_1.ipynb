This cell installs the pyspark library, which is necessary for working with Apache Spark in Python.

```python
# Import SparkSession
from pyspark.sql import SparkSession

# Create Spark Session
spark = SparkSession.builder.appName("EmployeeData").getOrCreate()
```

This cell imports SparkSession from pyspark.sql and then creates a SparkSession object named spark. The SparkSession is the entry point to programming Spark with the DataFrame and Dataset API.

```python
# Create data (20 records)
data = [
    (1, "Alice", 25, 25000),
    (2, "Bob", 30, 40000),
    (3, "Charlie", 28, 32000),
    (4, "David", 35, 28000),
    (5, "Eva", 26, 35000),
    (6, "Frank", 40, 50000),
    (7, "Grace", 29, 27000),
    (8, "Hannah", 32, 45000),
    (9, "Ian", 27, 29000),
    (10, "Jack", 31, 31000),
    (11, "Karen", 33, 38000),
    (12, "Leo", 36, 42000),
    (13, "Mona", 24, 26000),
    (14, "Nina", 29, 34000),
    (15, "Oscar", 41, 55000),
    (16, "Paul", 38, 36000),
    (17, "Quinn", 26, 23000),
    (18, "Rita", 34, 39000),
    (19, "Sam", 28, 30000),
    (20, "Tina", 37, 47000)
]
```

This cell defines a list of tuples, data, which will be used to populate the Spark DataFrame. Each tuple represents an employee record with an ID, Name, Age, and Salary.

```python
# Define columns
columns = ["ID", "Name", "Age", "Salary"]

# Create DataFrame
df = spark.createDataFrame(data, columns)
```

This cell defines the column names for the DataFrame as columns and then creates a Spark DataFrame named df using the data and columns defined in the previous cells.

```python
# Display Schema
print("Schema:")
df.printSchema()
```

This cell prints the schema of the created DataFrame df, showing the column names and their inferred data types.

```python
# Show all records
print("All Records:")
df.show()
```

This cell displays all the records contained within the df DataFrame in a tabular format, allowing for a quick overview of the data.

```python
# Filter employees with salary > 30000
print("Employees with Salary > 30000:")
filtered_df = df.filter(df.Salary > 30000)
filtered_df.show()
```

This cell filters the df DataFrame to select only those employees whose Salary is greater than 30000. The resulting filtered DataFrame is stored in filtered_df and then displayed.
